```r
library("coloc")
suppressMessages(suppressWarnings(library(tidyverse)))
library(locuscomparer)
```

```r
## Function
```

```r
qtl_gen <- function(eqtl_df,gwas_df,tissue="tissue"){

    eqtl_df <- eqtl_df%>%rename(SNP=rs_id_dbSNP151_GRCh38p7)
gwas <- gwas_df
input <- merge(eqtl_df, gwas, by="SNP", all=FALSE, suffixes=c("_eqtl","gwas"))   
result <- coloc.abf(dataset1=list(pvalues=input$p, type="quant", N=input$n,snp=input$SNP), dataset2=list(pvalues=input$pval_nominal, type="quant", N=nrow(eqtl_df),snp=input$SNP), MAF=input$maf)
a=(result$results%>%filter(SNP.PP.H4>0.80))$pvalues.df1
b_pip = (result$results%>%filter(SNP.PP.H4>0.80))$SNP.PP.H4
snp=(input%>%filter(p==a))$SNP
gwas_fn <- gwas%>%dplyr::select(SNP,p)%>%rename(rsid=SNP,pval=p)%>% distinct(rsid, .keep_all= TRUE)
eqtl_fn <- eqtl_df%>%dplyr::select(SNP,pval_nominal)%>%rename(rsid=SNP,pval=pval_nominal)%>% distinct(rsid, .keep_all= TRUE)
merged = merge(gwas_fn, eqtl_fn, by = "rsid", suffixes = c("1", "2"), all = FALSE)
merged=get_position(merged)
chr=unique(merged$chr)
population <- "EUR"
merged <- merged%>%mutate(logp1=-log10(pval1),logp2=-log10(pval2))%>%dplyr::select(rsid,logp1,logp2,chr,pos)
ld = retrieve_LD(chr, snp, population)
p <- make_combined_plot(merged,"gwas","eQTL", ld, chr=chr, snp,legend=TRUE,legend_position = 'bottomright',lz_ylab_linebreak = FALSE)
output_df <- input%>%filter(SNP==snp)%>%mutate(PIP=b_pip,Tissue=tissue)%>%select(Tissue,gene_id,gene_name,SNP,maf,pval_nominal,slope,slope_se,p,PIP)
return(output_df)
}
```

```r
## EQTL file list 
```

```r
df_eqtl <- list.files("../../ANALYSIS/eQTL/GTEx_Analysis_v8_eQTL/",pattern='',full.names = TRUE)
```

```r
## trans Z-score file list
```

```r
df_z <- list.files("../../ANALYSIS/SMR/trans_input/",pattern='*.ma',full.names = TRUE)
df_z <- df_z[ !grepl("smr", df_z) ]
```

```r
result <- list()
suppressMessages(suppressWarnings(for (f in seq_along(df_eqtl)){
 name <- tail(str_split(df_eqtl,"/|.v8")[[f]],n=2)[1]
    result_nested <- list()
    tryCatch({
    for (g in seq_along(df_z)){
    eqtl_df <- read.table(df_eqtl[f], header=T, as.is=T,fill=TRUE)
    gwas_df <- read.table(df_z[g], header=T, as.is=T)
    result_nested[[g]] <- qtl_gen(eqtl_df,gwas_df,name)
    }}, error=function(e){})
result[[f]] <- bind_rows(result_nested, .id = "column_label")
}))
```

```r
trans_df <- bind_rows(result, .id = "column_label")%>%select(-column_label)
write.table(trans_df,"../../ANALYSIS/COLOC/Trans_ancestry_coloc.tab",row.names=FALSE,quote=FALSE,sep="\t")
```
