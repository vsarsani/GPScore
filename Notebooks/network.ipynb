```r
genes <- c("CYP2R1","CSF1","LYPLAL1","IRS1","PBRM1","GNL3","ADIPOQ","RFC4","NAP1L5","FAM13A","FST","VEGFA","RGS17","TRIB1","ADRB1","PDE3A","RBMS2",
           "FGFR2","WDR11","RIC8B","CCDC92","DNAH10OS","ZNF664","CMIP","CDH13","CTDNEP1","ELP5","PEPD","CEBPG")
select_genes <- (mat%>%filter(Description %in% genes))$Name
```

```r
#WCGNA
```

```r
gene.names=rownames(oed)
select_index <- match(select_genes,gene.names)
```

```r
gene.names=rownames(oed)
trans.oed=t(oed)
```

```r
softPower = 6;

#calclute the adjacency matrix
adj= adjacency(trans.oed,type = "unsigned",selectCols = select_index, power = softPower);
```

```r
adj
```

```r
TOM=TOMsimilarityFromExpr(trans.oed,networkType = "unsigned", TOMType = "unsigned", power = softPower);
```

```r
##correlation
```

```r
corr_dat <- adj%>% as.data.frame %>% tibble::rownames_to_column() %>% 
    tidyr::pivot_longer(-rowname)%>%filter(value>0.95)%>%filter(rowname!=name)
```

```r
gene_ensgid <- mat%>%select(Name,Description)
corr_1 <- corr_dat%>%rename(Name=rowname)
merge_1 <- merge(corr_1,gene_ensgid,by="Name",all.x=TRUE)
corr_2<- merge_1%>%rename(Name1=Name,Name=name)
merge_2 <- merge(corr_2,gene_ensgid,by="Name",all.x=TRUE)
```

```r
final_df <- merge_2%>%select(Description.x,Description.y,value)%>%rename(GENE_1=Description.x,GENE_2=Description.y,correlation=value)
```
